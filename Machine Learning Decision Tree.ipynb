```python
import pandas as pd
import ast
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv("new.csv")
```

```python
print(df.groupby('Categories').size())
```

```python
df
```

```python
#count how many url exist in 'urls'
df['url_exist'] = df['urls'].apply(lambda x: ast.literal_eval(x))
df['url2'] = df['url_exist'].str.len()
#count how many 'photos' exist in a tweet
df['photos2'] = df["photos"].apply(lambda x: ast.literal_eval(x))
df['photos_exist'] = df['photos2'].str.len()

df['hashtags2'] = df['hashtags'].apply(lambda x: ast.literal_eval(x))
df['Length'] = df['hashtags2'].str.len()
df = df.rename(columns={'Length': 'hashtag_count', 'hashtags2': 'hashtag'})
```

```python
#Count average word per tweet
try:
    def avg_word(sentence):
        words = sentence.split()
        return (sum(len(word) for word in words)/len(words))
    df['avg_word']=df['tweet'].apply(lambda x: avg_word(x))
except ZeroDivisionError:
    print("There is a Zero Division")
```

```python
#collect dataframe on interest
df=df[['username','name','likes_count','replies_count','retweets_count','video','hashtag_count','url2','photos_exist', 'avg_word','Categories']]
```

```python
df.head(5)
```

```python
y = df.iloc[:,-1] #column at the end
X = df.iloc[:,2:9] #third column to 11st column
# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100) # 80% training and 20% test
```

```python
clf = DecisionTreeClassifier()

clf.fit(X_train, y_train)

## predicting the testing data
y_pred = clf.predict(X_test)
# print(y_pred)

### result of decision tree
conf = confusion_matrix(y_test, y_pred)
print(conf)

acc = accuracy_score(y_test, y_pred)
print(acc)

rep = classification_report(y_test, y_pred)
print(rep)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

model = RandomForestClassifier(n_estimators=100, 
                               bootstrap = True,
                               max_features = 'sqrt')


model.fit(X_train, y_train)
## predicting the testing data
y_pred = model.predict(X_test)
print(y_pred)


### result of decision tree
conf = confusion_matrix(y_test, y_pred)
print(conf)

acc = accuracy_score(y_test, y_pred)
print(acc)

rep = classification_report(y_test, y_pred)
print(rep)
```

```python
from sklearn import svm
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

# Create Linear SVM object
support = SVC()
param_grid = [{'kernel':['rbf'],'gamma':[50,5,10,0.5],
             'C':[10,0.1,0.001] }]
gsv = GridSearchCV(support,param_grid,cv=5,n_jobs=-1)
gsv.fit(X_train, y_train)
```

```python
print("Best HyperParameter: ",gsv.best_params_)
print("Best Accuracy: %.2f%%"%(gsv.best_score_*100))
```

```python
clf = SVC(C= 10, gamma = 0.5 )
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

### result of decision tree
conf = confusion_matrix(y_test, y_pred)
print(conf)

acc = accuracy_score(y_test, y_pred)
print(acc)

rep = classification_report(y_test, y_pred)
print(rep)
```

```python
import pickle
filename = 'finalized_model_num.sav'
pickle.dump(clf, open(filename, 'wb'))
```
